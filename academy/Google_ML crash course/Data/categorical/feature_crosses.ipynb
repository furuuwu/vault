<p><a href="/machine-learning/glossary#feature-cross"><strong>Feature crosses</strong></a> are created by
crossing (taking the Cartesian product of) two or more categorical or bucketed
features of the dataset. Like <a href="/machine-learning/crash-course/numerical-data/polynomial-transforms">polynomial
transforms</a>,
feature crosses allow linear models to handle nonlinearities. Feature crosses
also encode interactions between features.</p>

<p>For example, consider a leaf dataset with the categorical features:</p>

<ul>
<li><code translate="no" dir="ltr">edges</code>, containing values <code translate="no" dir="ltr">smooth</code>, <code translate="no" dir="ltr">toothed</code>, and <code translate="no" dir="ltr">lobed</code></li>
<li><code translate="no" dir="ltr">arrangement</code>, containing values <code translate="no" dir="ltr">opposite</code> and <code translate="no" dir="ltr">alternate</code></li>
</ul>

<p>Assume the order above is the order of the feature columns in a one-hot
representation, so that a leaf with <code translate="no" dir="ltr">smooth</code> edges and <code translate="no" dir="ltr">opposite</code> arrangement
is represented as <code translate="no" dir="ltr">{(1, 0, 0), (1, 0)}</code>.</p>

<p>The feature cross, or Cartesian product, of these two features would be:</p>

<p><code translate="no" dir="ltr">{Smooth_Opposite, Smooth_Alternate, Toothed_Opposite, Toothed_Alternate,
Lobed_Opposite, Lobed_Alternate}</code></p>

<p>where the value of each term is the product of the base feature values, such
that:</p>

<ul>
<li><code translate="no" dir="ltr">Smooth_Opposite = edges[0] * arrangement[0]</code></li>
<li><code translate="no" dir="ltr">Toothed_Opposite = edges[1] * arrangement[0]</code></li>
<li><code translate="no" dir="ltr">Lobed_Alternate = edges[2] * arrangement[1]</code></li>
</ul>

<p>For any given example in the dataset, the feature cross will equal 1 only if
both base features' original one-hot vectors were 1 for the crossed categories.
That is, an oak leaf with a lobed edge and alternate arrangement would have a
value of 1 only for <code translate="no" dir="ltr">Lobed_Alternate</code>, and the feature cross above would be:</p>

<p><code translate="no" dir="ltr">{0, 0, 0, 0, 0, 1}</code></p>

<p>This dataset could be used to classify leaves by tree species, since these
characteristics do not vary within a species.</p>

<p>
Feature crosses are somewhat analogous to

Polynomial transforms</a>.
Both combine multiple features into a new synthetic feature that the model can
train on to learn nonlinearities. Polynomial transforms typically combine
numerical data, while feature crosses combine categorical data.

<h2 id="when_to_use_feature_crosses" data-text="When to use feature crosses" tabindex="-1" role="presentation"><span class="devsite-heading" role="heading" aria-level="2">When to use feature crosses</span><button type="button" class="devsite-heading-link button-flat material-icons" aria-label="Copy link to this section: When to use feature crosses" data-title="Copy link to this section: When to use feature crosses" data-id="when_to_use_feature_crosses"></button></h2>

<p>Domain knowledge can suggest a useful combination of features
to cross. Without that domain knowledge, it can be difficult to determine
effective feature crosses or polynomial transforms by hand. It's often possible,
if computationally expensive, to use
<a href="/machine-learning/crash-course/neural-networks">neural networks</a> to
<em>automatically</em> find and apply useful feature combinations during training.</p>

<p>Be carefulâ€”crossing two sparse features produces an even sparser new
feature than the two original features. For example, if feature A is a
100-element sparse feature and feature B is a 200-element sparse feature,
a feature cross of A and B yields a 20,000-element sparse feature.</p>

Here are some examples that showcase the concept of feature crosses:

---

### **Example 1: Predicting House Prices**
Suppose you are trying to predict house prices using the following categorical features:
- **Neighborhood**: A categorical feature with values `Urban`, `Suburban`, `Rural`.
- **House Type**: A categorical feature with values `Detached`, `Semi-Detached`, `Apartment`.

#### Feature Cross:
The feature cross of `Neighborhood` and `House Type` would be:
```
{Urban_Detached, Urban_Semi-Detached, Urban_Apartment,
 Suburban_Detached, Suburban_Semi-Detached, Suburban_Apartment,
 Rural_Detached, Rural_Semi-Detached, Rural_Apartment}
```

- If a house is a **Detached** house in a **Suburban** neighborhood, the one-hot vector for the feature cross would be:
  ```
  {0, 0, 0, 1, 0, 0, 0, 0, 0}
  ```

This feature cross could help the model capture interactions between neighborhood type and house type that influence price. For example, **Suburban_Detached** houses might be priced higher than **Urban_Apartment** homes.

---

### **Example 2: Customer Churn Prediction**
In predicting whether a customer will leave a subscription service, you might have these features:
- **Age Group**: A categorical feature with values `Young`, `Middle-Aged`, `Senior`.
- **Subscription Plan**: A categorical feature with values `Basic`, `Standard`, `Premium`.

#### Feature Cross:
The feature cross of `Age Group` and `Subscription Plan` would be:
```
{Young_Basic, Young_Standard, Young_Premium,
 Middle-Aged_Basic, Middle-Aged_Standard, Middle-Aged_Premium,
 Senior_Basic, Senior_Standard, Senior_Premium}
```

- If a customer is **Middle-Aged** and on a **Standard** plan, their one-hot vector would be:
  ```
  {0, 0, 0, 0, 1, 0, 0, 0, 0}
  ```

This feature cross can help identify patterns such as **Senior_Premium** customers being less likely to churn compared to **Young_Basic** customers.

---

### **Example 3: Classifying User Interests**
For a recommendation system, you might have:
- **User Location**: A categorical feature with values `North America`, `Europe`, `Asia`.
- **Favorite Genre**: A categorical feature with values `Action`, `Drama`, `Comedy`.

#### Feature Cross:
The feature cross of `User Location` and `Favorite Genre` would be:
```
{NorthAmerica_Action, NorthAmerica_Drama, NorthAmerica_Comedy,
 Europe_Action, Europe_Drama, Europe_Comedy,
 Asia_Action, Asia_Drama, Asia_Comedy}
```

- A user from **Asia** who prefers **Drama** would have the one-hot vector:
  ```
  {0, 0, 0, 0, 0, 1, 0, 0, 0}
  ```

This cross might help the model recognize trends such as users from **Asia** having a higher preference for **Drama** movies.

---

### **Example 4: Diagnosing Medical Conditions**
In a medical dataset, consider:
- **Symptom 1**: A categorical feature with values `Fever`, `Cough`, `Fatigue`.
- **Symptom 2**: A categorical feature with values `Headache`, `Nausea`, `Dizziness`.

#### Feature Cross:
The feature cross of `Symptom 1` and `Symptom 2` would be:
```
{Fever_Headache, Fever_Nausea, Fever_Dizziness,
 Cough_Headache, Cough_Nausea, Cough_Dizziness,
 Fatigue_Headache, Fatigue_Nausea, Fatigue_Dizziness}
```

- A patient with **Cough** and **Dizziness** would have the one-hot vector:
  ```
  {0, 0, 0, 0, 0, 1, 0, 0, 0}
  ```

This feature cross could help identify unique symptom combinations associated with specific medical conditions.

---

### **Example 5: Predicting Flight Delays**
For predicting flight delays, you might have:
- **Airline**: A categorical feature with values `AirlineA`, `AirlineB`, `AirlineC`.
- **Day of Week**: A categorical feature with values `Monday`, `Tuesday`, ..., `Sunday`.

#### Feature Cross:
The feature cross of `Airline` and `Day of Week` would be:
```
{AirlineA_Monday, AirlineA_Tuesday, ..., AirlineA_Sunday,
 AirlineB_Monday, AirlineB_Tuesday, ..., AirlineB_Sunday,
 AirlineC_Monday, AirlineC_Tuesday, ..., AirlineC_Sunday}
```

- A flight with **AirlineB** scheduled for **Friday** would have the one-hot vector:
  ```
  {0, 0, ..., 0, 1, 0, ..., 0}
  ```

This feature cross could help the model capture interactions such as **AirlineC** flights being more likely to be delayed on weekends compared to weekdays.

---

In each example, the feature cross enables the model to capture interactions between the base features, which would be missed if the features were considered independently.
